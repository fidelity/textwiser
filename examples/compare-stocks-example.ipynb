```python
from blendsql import blend, SQLiteDBConnector, init_secrets
from blendsql.ingredients import LLMMap

init_secrets("../secrets.json")
```

```python
db_path = ""
db = SQLiteDBConnector(db_path)
# Give me some unique facts on the stocks in my portfolio
blendsql = """
SELECT Description, "Current Value" as "Total Value", Description,
{{
    LLMMap(
            'What are some facts an investor should know about this investment?',
            'portfolio::Description'
        )
}}
    FROM portfolio
    WHERE Description IN ('SEMPRA ENERGY', 'AMGEN', 'YUM! BRANDS', 'COMCAST')
"""
smoothie = blend(query=blendsql, db=db, ingredients={LLMMap}, verbose=False)
print(f"Passed {smoothie.meta.values_passed} values to our LLM function")
```
